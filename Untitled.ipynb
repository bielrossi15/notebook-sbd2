## Fine-tuning modelo BERT pre-treinado para avaliar se um tweet sera bem recebido ou não.

Aqui vamos considerar que um tweet é bem recebido se ele tem mais de 50 likes.

```python
df_bem_aceito = pd.DataFrame()
df_nao_aceito = pd.DataFrame()
```

```python
unique_bem_aceito = df[df["Likes"] > 50]["Tweet"].unique()
df_bem_aceito["Tweets"] = unique_bem_aceito
df_bem_aceito["labels"] = 1
```

```python
df_bem_aceito = df_bem_aceito.sample(frac=1, replace = False, random_state=20)
df_bem_aceito
```

```python
np.random.seed(20)
unique_nao_aceito = df[(df["RTs"] < 1) | (df["Likes"] < 1)]["Tweet"].unique()
unique_nao_aceito = np.random.choice(unique_nao_aceito, len(unique_bem_aceito), replace=False)
df_nao_aceito["Tweets"] = unique_nao_aceito
df_nao_aceito["labels"] = 0
```

```python
df_nao_aceito = df_nao_aceito.sample(frac=1, replace = False, random_state=20)
df_nao_aceito
```

```python
df_aceitacao = pd.concat([df_bem_aceito, df_nao_aceito]).sample(frac=1, random_state=20, replace = False).reset_index(0).drop(columns=['index'])
```

```python
df_aceitacao
```

```python
df_train_aceitacao = df_aceitacao[:1851]
df_val_aceitacao = df_aceitacao[1851:2082]
df_test_aceitacao = df_aceitacao[2082:]
```

```python
from datasets import Dataset, DatasetDict

dataset = DatasetDict({
    "train": Dataset.from_pandas(df_train_aceitacao),
    "test": Dataset.from_pandas(df_test_aceitacao),
    "eval": Dataset.from_pandas(df_val_aceitacao)
    })
```

```python
dataset
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained('bert-base-cased')
```

```python
def tokenize_data(dataset_to_tokenize):
    return tokenizer(dataset_to_tokenize['Tweets'], padding='max_length')

dataset = dataset.map(tokenize_data, batched=True)
```

```python
remove_columns = ['Tweets']
dataset = dataset.map(remove_columns=remove_columns)
dataset
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="tweetaceitacaomodel",
    learning_rate=4e-6,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=5,
    weight_decay=0.01,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    logging_steps = 10
)
```

```python
from transformers import AutoModelForSequenceClassification
model = AutoModelForSequenceClassification.from_pretrained("bert-base-cased", num_labels=2)
#model = AutoModelForSequenceClassification.from_pretrained("./tweetaceitacaomodel/checkpoint-232", num_labels=2)
```

```python
train_dataset = dataset['train']
eval_dataset = dataset['eval']
```

```python
from transformers import Trainer
from datasets import load_metric
metric = load_metric('accuracy')

def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=1)
    return metric.compute(predictions=predictions, references=labels)

trainer = Trainer(
    model=model, args=training_args, train_dataset=train_dataset, eval_dataset=eval_dataset, compute_metrics=compute_metrics
)
```

```python
trainer.train()
```

```python
import torch
torch.cuda.empty_cache()
```

```python
from sklearn.metrics import confusion_matrix
y_true = dataset["test"]["labels"]
y_pred = trainer.predict(dataset["test"].map(remove_columns=["labels"]))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

def greater(x):
  newl = []
  for i in x:
    if i[0] > i[1]:
      newl.append(0)
      continue
    newl.append(1)
  return newl

ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_true, greater(y_pred[0])), display_labels=[0, 1]).plot()
```

## Classificação de emoções usando BERT (cardiffnlp/twitter-roberta-base-emotion)

```python
from transformers import AutoModelForSequenceClassification
from transformers import TFAutoModelForSequenceClassification
from transformers import AutoTokenizer
import numpy as np
from scipy.special import softmax
import csv
import urllib.request

# Preprocess text (username and link placeholders)
def preprocess(text):
    new_text = []
    for t in text.split(" "):
        t = '@user' if t.startswith('@') and len(t) > 1 else t
        t = 'http' if t.startswith('http') else t
        new_text.append(t)
    return " ".join(new_text)

# Tasks:
# emoji, emotion, hate, irony, offensive, sentiment
# stance/abortion, stance/atheism, stance/climate, stance/feminist, stance/hillary

task='emotion'
MODEL = f"cardiffnlp/twitter-roberta-base-{task}"

tokenizer = AutoTokenizer.from_pretrained(MODEL)

# download label mapping
mapping_link = f"https://raw.githubusercontent.com/cardiffnlp/tweeteval/main/datasets/{task}/mapping.txt"
with urllib.request.urlopen(mapping_link) as f:
    html = f.read().decode('utf-8').split("\n")
    csvreader = csv.reader(html, delimiter='\t')
labels = [row[1] for row in csvreader if len(row) > 1]

# PT
model_emotion = AutoModelForSequenceClassification.from_pretrained(MODEL)
model_emotion.save_pretrained(MODEL)

```

```python
df["tokenized_bert"] = df["Tweet"].apply(lambda x: tokenizer(x, return_tensors='pt'))
```

```python
!pip install pandarallel
```

```python
from pandarallel import pandarallel

pandarallel.initialize(progress_bar=True, nb_workers=1)


def get_emotion(x):
  scores = x[0][0].detach().numpy()
  scores = softmax(scores)
  return labels[np.where(scores == np.max(scores))[0][0]]

df_10000 = df.sample(n=10000, random_state=20, replace = False).reset_index(0)

df_10000["emotions"] = df_10000["tokenized_bert"].parallel_apply(lambda x: get_emotion(model_emotion(**x)))
```
