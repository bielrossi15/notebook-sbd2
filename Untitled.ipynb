**Adicionando coluna de análise de sentimento**

```python
from textblob import TextBlob # for sentiment analysis

def sentiment_analysis(dataframe):
    dataframe['sentiment_percentage'] = [round(TextBlob(word).sentiment.polarity, 2) for word in dataframe['preprocessed_tweets']]
    dataframe['sentiment'] = ['positive' if percentage >= 0 else 'negative' for percentage in dataframe['sentiment_percentage']]
```

## Visualização dos dados

```python
plt.figure(figsize=(8, 8))
df['sentiment'].value_counts().plot(kind='pie')
```

**Criar wordcloud para valores de sentimento positivo e negativo**

```python
def create_positive_and_negative_array(dataframe):
    positive_values = np.where(dataframe['sentiment'] == 'positive', dataframe['preprocessed_tweets'] , 0)
    negative_values = np.where(dataframe['sentiment'] == 'negative', dataframe['preprocessed_tweets'] , 0)
    
    positive_values = [value for value in positive_values if value != 0]
    negative_values = [value for value in negative_values if value != 0]
    
    return positive_values, negative_values

positive_values, negative_values = create_positive_and_negative_array(df)
```

```python
from wordcloud import WordCloud

wordcloud = WordCloud(background_color="white",width=1600, height=800).generate(' '.join(positive_values))
plt.figure( figsize=(20,10), facecolor='k')
plt.imshow(wordcloud)
plt.axis('off')
```

```python
wordcloud = WordCloud(background_color="white",width=1600, height=800).generate(' '.join(negative_values))
plt.figure( figsize=(20,10), facecolor='k')
plt.imshow(wordcloud)
plt.axis('off')
```

**Matriz de correlação, para entender como os valores numéricos estão relacionados**

```python
def correlation_heatmap(dataframe, is_triangle):
    plt.figure(figsize=(15, 10))
    if is_triangle:
        mask = np.triu(np.ones_like(dataframe.corr()))
        sns.heatmap(dataframe.corr(), mask=mask, annot=True, cmap='BrBG')
    else:
        sns.heatmap(dataframe.corr(), annot=True, cmap='YlGnBu')
```

```python
correlation_heatmap(df, False)
```

```python
correlation_heatmap(df, True)
```
